```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Gastos y costos 20-22_limpio (Hoja 2020)**

```python
df2020 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2020", skiprows=5)
df2020.head(3)
```

```python
#Hacer una copia de la base de datos original
df_2020 = df2020.copy()
```

```python
#Identificar valores nulos por columna
valores_nulos=df_2020.isnull().sum()
valores_nulos
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
df_2020.drop(['Columna1'], axis=1)
```

```python
#Identificar valores nulos por columna
valores_nulos=df_2020.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con promedio o media -IMPORTE-
df_2020["IMPORTE"]=df_2020["IMPORTE"].fillna(round(df_2020["IMPORTE"].mean(),1))
```

```python
#Identificar valores nulos por columna
valores_nulos=df_2020.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un string en concreto
df_2020["GASTO"]=df_2020["GASTO"].fillna("No aplica")
```

```python
#Sustituir valores nulos por un valor numérico en  concreto -TC-
df_2020["TC"]= df_2020["TC"].fillna(0)
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") -FOLIO-
#Filtro por columnas
df_2020["FOLIO"] =df_2020["FOLIO"].fillna(method="ffill")
```

```python
#Sustituir valores nulos por un valor numérico en  concreto -IVA-
df_2020["IVA"]= df_2020["IVA"].fillna(0)
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill") -TIPO-
#Filtro por columnas
df_2020["TIPO"] =df_2020["TIPO"].fillna(method="ffill")
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante "forward fill" ("ffill") -POLIZA-
# Convertir la columna a tipo texto, valores no convertibles se vuelven NaN
columna = ["POLIZA"]
df_2020[columna] = df_2020[columna].astype(str)
#Filtro por columnas
df_2020["POLIZA"] = df_2020["POLIZA"].fillna(method="ffill")
```

```python
#Identificar valores nulos por columna
df_2020.isnull().sum()
```

```python
#Convertir DataFrame a CSV
df_2020.to_csv("Gastos&Costos_2020.csv")
```

**Gastos y costos 20-22_limpio (Hoja 2021)**

```python
df2021 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2021", skiprows=3)
df2021.head(3)
```

```python
#Hacer una copia del dataframe original
df_2021 = df2021.copy()
```

```python
#Identificar valores nulos por columna
df_2021.isnull().sum()
```

```python
#Indentificar valores nulos por dataframe
df_2021.isnull().sum().sum()
```

```python
#Identificar el número de datos y columnas
df_2021.shape
```

```python
#Identificar el tipo de datos de las columnas que tienen valores nulos
df_2021.dtypes[df_2021.isnull().any()]
```

```python
# Convertir columnas a tipo texto, valores no convertibles se vuelven NaN
columnas = ["FOLIO","MP", "POLIZA"]
df_2021[columnas] = df_2021[columnas].astype(str)
```

```python
#Conocimeinto de la columna 'FOLIO'(147 valores nulos).
#Número de valores únicos en la columna:
print(df_2021['FOLIO'].nunique())
# Lista de valores únicos
print(df_2021['FOLIO'].unique())

#Verificar si existe este folio 000
existe = "000" in df_2021["FOLIO"].values
print(existe)

#Sustituir los NA's con este folio para saber que se desconoce
df_2021["FOLIO"] = df_2021["FOLIO"].fillna("000")
```

```python
#Conocimeinto de la columna 'POLIZA'(2372 valores nulos).
#Número de valores únicos en la columna:
print(df_2021['POLIZA'].nunique())

#Sustituir los NA's leyenda "SIN INFORMACIÓN"
df_2021["POLIZA"] = df_2021["POLIZA"].fillna("SIN INFROMACIÓN")
```

```python
#Conocimiento de la columna 'MP' (654 valores nulos)
#Número de valores únicos en la columna:
print(df_2021['MP'].nunique())
# Lista de valores únicos
print(df_2021['MP'].unique())

#Cuantos valores hay de cada valor único
conteo = df_2021['MP'].value_counts()
print(conteo)

#Saber cuantos proveedores hay de cada MP
resumen = df_2021.groupby(["PROVEEDOR", "MP"]).size().unstack(fill_value=0)
print(resumen)

#Sustituir los NA'S con los valores mayores de los proveedores dados de alta
# Paso 1: calcular el valor mayoritario de PUE/PPD por proveedor
mayoria = df_2021.groupby("PROVEEDOR")["MP"].apply(lambda x: x.dropna().mode()[0] if not x.dropna().empty else None)

# Paso 2: usar ese valor para rellenar los NaN en MP
df_2021["MP"] = df_2021.apply(
    lambda row: mayoria[row["PROVEEDOR"]] if pd.isna(row["MP"]) else row["MP"],
    axis=1
)
```

```python
#Verificación de NA's y exportación de dataframe
df_2021.isnull().sum().sum()
df_2021.to_csv("Gastos&Costos_2021.csv")
```

**Gastos y costos 20-22_limpio (Hoja 2022)**

```python
df2022= pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name= "2022",  skiprows=3)
print(df2022.shape)
df2022.head(10)
```

```python
valores_nulos=df2022.isnull().sum()
valores_nulos
```

```python
#Crear una copia del df original
df_2022 = df2022.copy()
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df_2022["Folio"] = df_2022["Folio"].fillna(method="ffill")
```

```python
#Sustitución de valores nulos de acuerdo a condicional de columna Descripción y MP (Pago = PAGO)
MP_NULL = df_2022['MP'].isna()
df_2022.loc[MP_NULL, 'MP'] = 'PAGO'
```

```python
# Rellenar valores nulos de TC con 0
df_2022["TC"] = df_2022["TC"].fillna(0)
```

```python
# Rellenar valores nulos de Poliza con "SIN POLIZA"
df_2022["Poliza"] = df_2022["Poliza"].fillna("SIN POLIZA")
```

```python
# Eliminar la columna "Otros"
df_2022 = df_2022.drop(columns=["Otros"])
```

```python
#Verificación de valores nulos en el df
df_2022.isnull().sum()
```

```python
#Exportación de base de datos limpio a csv
df_2022.to_csv("Gastos&Costos_2022.csv")
```

**Gastos y costos 20-23_limpio (Hoja 2023)**

```python
# Carga de archivo
df2023 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2023", skiprows=3)
print(df2023.columns)
print(df2023.shape)
df2023.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=df2023.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
df_2023 = df2023
df_2023.head(5)
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df_2023["Folio"] = df_2023["Folio"].fillna(method="ffill")
df_2023["Proveedor"] = df_2023["Proveedor"].fillna(method="ffill")
df_2023["TIPO GASTO"] = df_2023["TIPO GASTO"].fillna(method="ffill")
```

```python
#Sustitución de valores nulos de acuerdo a condicional de columna Descripción y MP (Pago = PAGO)
MP_NULL = df_2023['MP'].isna()
df_2023.loc[MP_NULL, 'MP'] = 'PAGO'

#Sustitución de valores nulos de acuerdo a condicional de columna Descripción y FP (Pago = 00)
df_2023["FP"]= df_2023["FP"].fillna(00)
df_2023

display(df_2023.isnull().sum())
```

```python
#Eliminar columna completa 'Otros' sin valores
df_2023 = df_2023.drop(["Otros"], axis = 1)
df_2023
```

```python
#Rellenar los valores nulos con nombre 'NONE' en la variable 'Poliza'
df_2023['Poliza'] = df_2023['Poliza'].fillna('NONE')
display(df_2023.isnull().sum())
```

```python
#Pasar dataframe limpio a csv.
df_2023.to_csv("Gastos&Costos_2023.csv")
```
