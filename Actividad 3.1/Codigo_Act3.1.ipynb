```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga de datos
ventas = pd.read_csv('Ventas_totales.csv')
ventas.head(5)
```

```python
#Indentificar valores nulos por dataframe 
ventas.isnull().sum().sum()
```

APLICACIÃ“N DE METODOS DE SUSTITUCIÃ“N EN LOS VALORES NULOS PRESENTADOS

```python
#RealizaciÃ³n de copia de dataframe
#Se realiza la copia del dataframe original para no alterar los datos originales, ademÃ¡s de poder comparar resultados al final del anÃ¡lisis
df = ventas.copy()
```

     APLICACIÃ“N DE MÃ‰TODO DE SUSTITUCIÃ“N DE VALORES POR MEDIA

```python
#Se utiliza un grÃ¡fico de lÃ­neas para observar los valores en las columnas y detectar posibles outliers

#Columna otros_medios
plt.figure(figsize=(10,3.5))
plt.plot(df['otros_medios'], marker='*', linestyle='-', alpha=0.7, color='steelblue')
plt.title("Valores de pago con otros medios")
plt.grid(True)
plt.show()


#Columna panderia
plt.figure(figsize=(10,3.5))
plt.plot(df['panaderia'], marker='o', linestyle='--', alpha=0.7, color='dodgerblue')
plt.title("Valores en panaderia")
plt.grid(True)
plt.show()


#Columna lacteos
plt.figure(figsize=(10,3.5))
plt.plot(df['lacteos'], marker='^', linestyle='-', alpha=0.7, color='dodgerblue')
plt.title("Valores en lÃ¡cteos")
plt.grid(True)
plt.show()

```

```python
#Remplazo de valores nulos por la media de las sigueitnes columnas: otros_medios (1 unico valor nulo), panaderia (1 unico valor nulo)
#se eligio la media gracias a que imputar con ella evita que los nulos distorsionen la magnitud representativa de la columna.  
ventas[['otros_medios', 'panaderia','lacteos',
        'carnes']]=ventas[['otros_medios', 'panaderia', 'lacteos',
                           'carnes']].fillna(ventas[['otros_medios','panaderia','lacteos','carnes']].mean())
```

     APLICACIÃ“N DE MÃ‰TODO DE SUSTITUCIÃ“N DE VALORES POR BFILL & FFILL

```python
#GrÃ¡fico lÃ­neal para observaciÃ³n en las columnas y detectar posibles outliers

#Columna otros
plt.figure(figsize=(10,3.5))
plt.plot(df['otros'], marker='o', linestyle='--', alpha=0.7, color='firebrick')
plt.title("Valores de columna otros")
plt.grid(True)
plt.show() 

#Columna tarjetas_debito
plt.figure(figsize=(10,3.5))
plt.plot(df['tarjetas_debito'], marker='*', linestyle='--', alpha=0.7, color='salmon')
plt.title("Valores de pago con tarjeta de debito")
plt.grid(True)
plt.show()

#   _____ 


#Columna almacen
plt.figure(figsize=(10,3.5))
plt.plot(df['almacen'], marker='o', linestyle='--', alpha=0.7, color='darkred')
plt.title("Valores en almacÃ©n")
plt.grid(True)
plt.show() 

#Columna electronicos_articulos_hogar
plt.figure(figsize=(10,3.5))
plt.plot(df['electronicos_articulos_hogar'], marker='^', linestyle='-', alpha=0.7, color='firebrick')
plt.title("Valores en electronicos y articulos del hogar")
plt.grid(True)
plt.show()

```

```python
#Remplazo de valores nulos por el mÃ©todo backward fill a las columnas: otros (4 valres nulos), tarjeta_debito (1 valor nulo)
#Se utiliza este metodo gracias a como podemos osbervar en los grÃ¡ficos de lÃ­neas que los valores anteriores pueden persivirse como outliers, por lo que se decide no utilizarlos para el relleno de los valores nulos, 
#en cambio se utilizan los valores posteriores para el relleno de los valores nulos
ventas[['otros', 'tarjetas_debito']]=ventas[['otros', 'tarjetas_debito']].fillna(method='bfill')


#Remplazo de valores nulos por la mÃ©todo forward fill a las columnas: almacen (2 valores nulo), electronicos_articulos_hogar (3 valores nulos)
#Se utiliza este metodo gracias a como podemos osbervar en los grÃ¡ficos de lÃ­neas que los valores anteriores no pueden persivirse como outliers, por lo que se decide no utilizarlos para el relleno de los valores nulos,
ventas[['almacen', 'electronicos_articulos_hogar']]=ventas[['almacen', 'electronicos_articulos_hogar']].fillna(method='ffill')
```

     APLICACIÃ“N DE MÃ‰TODO DE SUSTITUCIÃ“N DE VALORES POR MEDIANA

```python
#Se utiliza un grÃ¡fico de lÃ­neas para observar los valores en la columna y detectar posibles outliers

#Columna tarjetas_credito
plt.figure(figsize=(10,3.5))
plt.plot(df['tarjetas_credito'], marker='^', linestyle='-', alpha=0.7, color='indigo')
plt.title("Valores de pago en tarjeta de crÃ©dito")
plt.grid(True)
plt.show() 

#Columna subtotal_ventas_alimentos_bebidas
plt.figure(figsize=(10,3.5))
plt.plot(df['subtotal_ventas_alimentos_bebidas'], marker='o', linestyle='-', alpha=0.7, color='blueviolet')
plt.title("Valores de los subtotales en ventas de alimentos y bebidas")
plt.grid(True)
plt.show() 

#Columna salon_ventas
plt.figure(figsize=(10,3.5))
plt.plot(df['salon_ventas'], marker='o', linestyle='--', alpha=0.7, color='darkmagenta')
plt.title("Valores de ventas de los salones")
plt.grid(True)
plt.show()

#Columna bebidas
plt.figure(figsize=(10,3.5))
plt.plot(df['bebidas'], marker='*', linestyle='-', alpha=0.7, color='darkviolet')
plt.title("Valores en las bebidas")
plt.grid(True)
plt.show()

#Columna carnes
plt.figure(figsize=(10,3.5))
plt.plot(df['carnes'], marker='o', linestyle='-', alpha=0.7, color='purple')
plt.title("Valores en carnes")
plt.grid(True)
plt.show()

#Columna indumentaria_calzado_textiles_hogar
plt.figure(figsize=(10,3.5))
plt.plot(df['indumentaria_calzado_textiles_hogar'], marker='^', linestyle='-', alpha=0.7, color='rebeccapurple')
plt.title("Valores en la indumentaria del calzado y los textiles del hogar")
plt.grid(True)
plt.show()
```

```python
#Se renplazan los valores nulos en las siguientes columnas:subtotal_ventas_alimentos_bebidas (10 valores nulos), columna tarjetas_credito (7 valores nulos), salones_ventas (6 valores nulos), 
#                                                           bebidas (1 valor nulo), carnes (1 valor nulo), indumentaria_calzado_textiles_hogar (2 valores nulos), 
#se utiliza la mediana para el relleno de los valores nulos, ya que en el grÃ¡fico se observan posibles outliers que podrÃ­an distorsionar la media o un relleno por mÃ©todo forward o backward fill
ventas[['subtotal_ventas_alimentos_bebidas','tarjetas_credito', 'bebidas', 'carnes', 'indumentaria_calzado_textiles_hogar', 'salon_ventas']]=ventas[['subtotal_ventas_alimentos_bebidas', 'tarjetas_credito', 'indumentaria_calzado_textiles_hogar','bebidas','carnes','salon_ventas']].fillna(ventas[['subtotal_ventas_alimentos_bebidas', 'tarjetas_credito', 'bebidas', 'carnes', 'indumentaria_calzado_textiles_hogar', 'salon_ventas']].median())    
```

     APLICACIÃ“N DE MÃ‰TODO DE SUSTITUCIÃ“N DE VALORES POR VALORES NUMÃ‰RICOS EN CONCRETO

```python
#Se utiliza un grÃ¡fico de lÃ­neas para observar los valores en la columna y detectar posibles outliers

#Columna verduleria_fruteria
plt.figure(figsize=(10,3.5))
plt.plot(df['verduleria_fruteria'], marker='^', linestyle='-', alpha=0.7, color='gold')
plt.title("Valores en veluleria y fruteria")
plt.grid(True)
plt.show() 

#Columna alimentos_preparados_rotiseria
plt.figure(figsize=(10,3.5))
plt.plot(df['alimentos_preparados_rotiseria'], marker='o', linestyle='-', alpha=0.7, color='orange')
plt.title("Valores en alimentos preparados y rostiseria")
plt.grid(True)
plt.show() 
```

```python
#Para no dejar a un lado las distinos metodos de sustituciÃ³n de valores nulos con el promedio de las siguientes columnas: alimentos_preparados_rotiseria (1 unico valor nulo), verduleria_fruteria (1 unico valor nulo)
df["verduleria_fruteria"].mean()
df["alimentos_preparados_rotiseria"].mean()
                                                                                                
ventas['verduleria_fruteria']= ventas['verduleria_fruteria'].fillna(2644027.3529390623) 
ventas['alimentos_preparados_rotiseria']= ventas['alimentos_preparados_rotiseria'].fillna(689586.284725)
```
